```python
import nibabel as nib
import matplotlib.pyplot as plt

anat_img = nib.load('../ADNI3/seg/ADNI_011_S_6303_MR_Accelerated_Sagittal_MPRAGE__br_raw_20190430142811025_189_S819896_I1160021.nii')
anat_img_data = anat_img.get_fdata()
print(anat_img_data.shape)

def show_slices(slices):
    """ Function to display row of image slices """
    fig, axes = plt.subplots(1, len(slices))
    for i, slice in enumerate(slices):
        axes[i].imshow(slice.T, cmap="gray", origin="lower")

show_slices([anat_img_data[91, :, :],
            anat_img_data[:, 109, :],
            anat_img_data[:, :, 91]])
```
